```python
import pandas as pd
df = pd.read_csv("kelas_era.csv")
display(df)
```

***Pembagian data latih dan uji 70:30***

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['komentar'], df['kelas'],
                                                    test_size = 0.30,
                                                    random_state = 0)
```

**FITUR EXTRACTION DENGAN TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
tfidf_train = tfidf_vectorizer.fit_transform(X_train)
tfidf_test = tfidf_vectorizer.transform(X_test)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectorizer.fit(X_train)
```

```python
X_train = vectorizer.transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
print(X_train)
```

**ALGORITMA SUPPORT VECTOR MACHINE**

```python
from sklearn import metrics
from sklearn.svm import SVC

svm = SVC(C=1.0, kernel='linear', degree=3, gamma='auto')
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
svm_akurasi = metrics.accuracy_score(y_test, y_pred)
print("Akurasi SVM:", svm_akurasi)
```

**EVALUASI MODEL MENGGUNAKAN CONFUSION MATRIX**

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix:")
print(cm)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, cmap="YlGnBu", fmt="d")
plt.xlabel("Prediksi")
plt.ylabel("Aktual")
plt.title("Confusion Matrix")

plt.show()
```

```python
print(classification_report(y_test, y_pred))
```
