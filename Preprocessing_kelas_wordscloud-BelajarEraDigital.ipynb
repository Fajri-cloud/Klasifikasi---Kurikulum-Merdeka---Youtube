**#PREPROCESSING**

***Cleaning***

```python
import re
import pandas as pd
def cleaning(text):
  if not isinstance(text, (str, bytes)):
    return text
  emoji_pattern = re.compile("["
      u"\U0001F600-\U0001F64F"
      u"\U0001F300-\U0001F5FF"
      u"\U0001F680-\U0001F6FF"
      u"\U0001F700-\U0001F77F"
      u"\U0001F780-\U0001F7FF"
      u"\U0001F800-\U0001F8FF"
      u"\U0001F900-\U0001F9FF"
      u"\U0001FA00-\U0001FA6F"
      u"\U0001FA70-\U0001FAFF"
      u"\U0001F000-\U0001F0FF"
      u"\U0001F200-\U0001F2FF"
      u"\U0001F400-\U0001F4FF"
      "]+", flags=re.UNICODE)
  text = emoji_pattern.sub(r'', text)
  text = re.sub(r'\d+', '', text)
  text = re.sub(r'[^\w\s]', '', text)
  for key, value in singkatan.items():
    text = text.replace(key, value)
  return text

df['cleaning'] = df['komentar'].apply(cleaning)
df.head(10)
```

***Case Folding***

```python
df['case folding'] = df['cleaning'].str.lower()
df.head(10)
```

***Tokenize***

```python
import nltk
import pandas as pd

nltk.download('punkt')

def tokenize(text):
    if isinstance(text, str):
        return nltk.word_tokenize(text)
    else:
        return []

df['tokenize'] = df['case folding'].apply(tokenize)
df.head(10)
```

***Stopword Removal***

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords = stopwords.words('indonesian')

def remove_stopwords(text):
    if isinstance(text, list):
        return [word for word in text if word not in stopwords]
    else:
        return []
df['stopwords removal'] = df['tokenize'].apply(remove_stopwords)
df.head(10)
```

***Normalize***

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.tokenize import word_tokenize

factory = StemmerFactory()
stemmer = factory.create_stemmer()
def normalize(text):
    if isinstance(text, list):
        return [stemmer.stem(word) for word in text]
    else:
        return stemmer.stem(text)
        return[]
df['normalize'] = df['stopwords removal'].apply(normalize)
df.head(5)
```

***Menghapus kolom dan baris bersifat kosong atau NaN***

```python
df.dropna(subset=['komentar'], inplace=True)
df.dropna(subset=['cleaning'], inplace=True)
df.dropna(subset=['case folding'], inplace=True)
df.dropna(subset=['stopwords removal'], inplace=True)
df.dropna(subset=['tokenize'], inplace=True)
df.dropna(subset=['normalize'], inplace=True)

df.head(10)
```

**TAHAP PELABELAN**

***Melakukan Tahapan pelabelan kelas menjadi Positif dan Negatif menggunakan Teknik Lexicon Based***

```python
!pip install vaderSentiment
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
df['nilai'] = df['normalize'].apply(lambda x: analyzer.polarity_scores(' '.join(x))['compound'])
df.head(10)
```

```python
df['kelas'] = df['nilai'].apply(lambda x: 'positif' if x > 0.5 else 'negatif')
df.head(5)
```

***Melihat Jumlah kelas positif dan negatif***

```python
positif_count = df['kelas'].value_counts()['positif']
negatif_count = df['kelas'].value_counts()['negatif']

print(f"Jumlah komentar positif: {positif_count}")
print(f"Jumlah komentar negatif: {negatif_count}")
```

```python
import matplotlib.pyplot as plt

labels = ['Positif', 'Negatif']
values = [positif_count, negatif_count]
colors = ['skyblue', 'red']

plt.bar(labels, values, color=colors)

total = sum(values)
for i, value in enumerate(values):
  percentage = value / total * 100
  plt.text(i, value, f'{percentage:.1f}%', ha='center', va='bottom')

plt.title('Perbandingan Komentar Positif dan Negatif')
plt.xlabel('Kelas')
plt.ylabel('Jumlah Komentar')

plt.show()
```
