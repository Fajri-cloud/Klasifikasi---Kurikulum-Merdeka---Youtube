```python
from googleapiclient.discovery import build
import pandas as pd

api_key =  'AIzaSyCedR4YohLFo7fsfhAhSPWeldZQy3OhDAE'

youtube = build('youtube', 'v3', developerKey=api_key)

def get_video_comments(video_id):
    comments = []
    response = youtube.commentThreads().list(
        part='snippet',
        videoId=video_id,
        textFormat='plainText'
    ).execute()

    while response:
        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']
            comments.append({
                'nama pengguna': comment['authorDisplayName'],
                'tanggal dan tahun': comment['publishedAt'],
                'komentar': comment['textDisplay'],
            })
        if 'nextPageToken' in response:
            response = youtube.commentThreads().list(
                part='snippet',
                videoId=video_id,
                textFormat='plainText',
                pageToken=response['nextPageToken']
            ).execute()
        else:
            break
    return comments

video_id = 'PphBLIh_jgU'
comments = get_video_comments(video_id)

df = pd.DataFrame(comments)
df.to_csv('belajareradigital.csv', index=False)
display(df)
```
